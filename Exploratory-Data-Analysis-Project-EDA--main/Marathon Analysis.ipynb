```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('TWO_CENTURIES_OF_UM_RACES.csv')
df.head(6)
```

```python
df.shape
```

```python
#cleaning data
#getting USA races, 50km or 50 mi, 2020

#combine 50km and 50mi - 2020 
df[(df['Event distance/length'].isin(['50km','50mi'])) & (df['Year of event'] == 2020) ]
```

```python
#filter USA

df[df['Event name'].str.split('(').str.get(1).str.split(')').str.get(0) == 'USA']
```

* .str.split('('): This splits the string 'Conference (USA)' into substrings at '(' character, resulting in a list of substrings: ['Conference ', 'USA)'].
* 
.str.get(1): This selects the second element (index 1) from the list obtained in step 1, which is 'USA)'.* 
.str.split(')'): This splits the string 'USA)' into substrings at ')' character, resulting in a list of substrings: ['USA', '']* .
.str.get(0): This selects the first element (index 0) from the list obtained in step 3, which is 'USA'.

```python
#combining all distances, year and country

df[(df['Event distance/length'].isin(['50km','50mi'])) & (df['Year of event'] == 2020) & (df['Event name'].str.split('(').str.get(1).str.split(')').str.get(0) == 'USA')]
```

```python
df2 = df[(df['Event distance/length'].isin(['50km','50mi'])) & (df['Year of event'] == 2020) & (df['Event name'].str.split('(').str.get(1).str.split(')').str.get(0) == 'USA')]
```

```python
#remove USA
df2['Event name'] = df2['Event name'].str.split('(').str.get(0)
```

```python
#clean up age

df2['athlete age'] = 2023 - df2['Athlete year of birth']
```

```python
#remove h from performance

df2['Athlete performance'] = df2['Athlete performance'].str.split(' ').str.get(0)
```

```python
#drop columns - club, country, YOB, age category

df2 = df2.drop(['Athlete club','Athlete country','Athlete age category', 'Athlete year of birth'],axis = 1)
```

```python
#rename columns
df2 = df2.rename(columns = {
                'Year of event': 'year',
                'Event dates': 'race_day',
                 'Event name': 'race_name', 
                 'Event distance/length': 'race_length', 
                 'Event number of finishers': 'race_no_of_finishers' ,
                 'Athlete performance': 'athlete_performance', 
                 'Athlete gender': 'athlete_gender',
                 'Athlete average speed': 'athlete_average_speed', 
                 'Athlete ID': 'athlete_id', 
                 'athlete age':'athlete_age' 
})
df2
```

```python
#reorder columns

df3 = df2[['race_day','race_name','race_length', 'athlete_id', 'athlete_gender','athlete_age', 'athlete_performance','athlete_average_speed', 'race_no_of_finishers','year']]
```

```python
sns.histplot(df3['race_length'])
plt
```

```python
sns.lmplot(data=df3, x = 'athlete_age', y = 'athlete_average_speed')

plt.title('Athlete Avg Speed Vs Athlete Age', color = 'purple')
plt.xlabel('Athlete Age')
plt.ylabel('Athlete Avg Speed')
plt.show
```

```python
sns.lmplot(data=df3, x = 'athlete_age', y = 'athlete_average_speed',hue = 'athlete_gender')

plt.title('Athlete Avg Speed Vs Athlete Age & Gender', color = 'red')
plt.xlabel('Athlete Age')
plt.ylabel('Athlete Avg Speed')
plt.show
```
